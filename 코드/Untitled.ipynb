```python
import os

# Getting all the arff files from the current directory
files = [arff for arff in os.listdir('.') if arff.endswith(".arff")]

# Function for converting arff list to csv list
def toCsv(content):
    data = False
    header = ""
    newContent = []
    for line in content:
        if not data:
            if "@attribute" in line:
                attri = line.split()
                columnName = attri[attri.index("@attribute")+1]
                header = header + columnName + ","
            elif "@data" in line:
                data = True
                header = header[:-1]
                header += '\n'
                newContent.append(header)
        else:
            newContent.append(line)
    return newContent

# Main loop for reading and writing files
for file in files:
    with open(file , "r") as inFile:
        content = inFile.readlines()
        name,ext = os.path.splitext(inFile.name)
        new = toCsv(content)
        with open(name+".csv", "w") as outFile:
            outFile.writelines(new)
```

```python
def column(a_list, num):
    if num == 0:#number of row가 바뀌지 않을 것
        print(a_list[0][1], "~", a_list[0][-1], "사이의 숫자를 입력 해 주세요.")
        return
    if num <= -1:#column의 범위보다 underflow인 숫자를 입력 시 일반 list와 똑같이 기능하도록 함
        num = a_list[0][num]
        
    attList = []
    print("Attribute of",num)
    for i in range(len(a_list)-1):#헤더 부분을 출력하지 않기 위해 -1
        attList.append(a_list[i+1][num])
        print(attList[i])
        
    return attList
```

```python
def columnReverse(a_list, num): 
    if num == 0:#number of row가 바뀌지 않을 것
        print(a_list[0][1], "~", a_list[0][-1], "사이의 숫자를 입력 해 주세요.")
        return
    if num <= -1:#column의 범위보다 underflow인 숫자를 입력 시 일반 list와 똑같이 기능하도록 함
        num = a_list[0][num]
        
    attList = []
    tempList = []
    for i in range(len(a_list)-1):#헤더 부분을 출력하지 않기 위해 -1
        attList.append(a_list[i+1][num])
    attList.reverse()
    
    for i in range(len(attList)):
        tempList.append(a_list[i+1][num+1:])#헤더와 바뀌는 열은 제외하고 임시로 저장
        del(a_list[i+1][num:])#바뀌는 열 부터 마지막 열 까지 삭제
    
    for i in range(len(attList)):
        a_list[i+1].append(attList[i])
        a_list[i+1].extend(tempList[i])#list 형태를 append 하게 되면 list형태로 추가되기 때문에 내부 요소만 추가.
    
    #for i in range(len(a_list)):
        #print(a_list[i])
    
    return a_list
```

```python
def divide_train_test(in_list, prop):
    num = in_list[-1][0]#마지막 행의 번호를 읽어옴
    trainNum = int(num*prop)+1#header 제외
    train_data = in_list[1:trainNum]#header+1 : trainNum
    test_data = in_list[trainNum:]#trainNum : -1
    
    return train_data, test_data
```

```python
def equ_discretize(_range, intervals):
    output = []
    _min, _max = _range[0], _range[1]
    a_min, a_max=  abs(_min), abs(_max)#절댓값
    distance = a_min + a_max
    i_distance = int(distance/intervals)# 1회 당 거리
    for i in range(intervals):
        output.append([_min, _min+i_distance])
        _min += i_distance
    return output
```

```python
equ_discretize([-4,8],3)
```
